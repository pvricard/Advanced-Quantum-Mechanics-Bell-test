```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#hidden protocol
def f(n,theta):
  x=2*np.pi*np.random.rand(n)
  flag=False
  if theta>np.pi:
    flag=True
    theta=theta-np.pi
  correlated1=((x<np.pi)&(x >= theta)).sum()
  correlated2=((np.pi<x)&( x >= theta+np.pi)).sum()
  correlated11=((x<np.pi)&(x <= theta)).sum()
  correlated22=((np.pi<x)&( x <= theta+np.pi)).sum()
  if flag==True:
    return ((+correlated1-correlated11+correlated2-correlated22)/n)
  else:
    return ((-correlated1+correlated11-correlated2+correlated22)/n)
```

```python
#statistic qm
def q11(theta2):
  x=-np.cos(theta2)
  return x
```

```python
N=1000
x=np.linspace(0,2*np.pi,N)
n=100000
yhidden=[]
theta_state=2*np.pi*np.random.rand()
```

```python
for theta in x:
  yhidden.append(f(n,theta))
ystat=q11(x)
```

```python
stepsx = np.arange(0, 2 * np.pi+np.pi/4, step=(np.pi / 4))
labelsx=[0]
for i in range(len(stepsx)-1):
    if (i+1)%2!=0:
        label=str(i+1)+"π/4"
        if (i+1)==1:
                label='π/4'
    else:
        labeli=(i+1)
        if labeli%4==0:
            label=str(int(labeli/4))+'π'
            if labeli/4==1:
                label='π'
        else:
            label=str(int(labeli/2))+'π/2'
            if labeli/2==1:
                label='π/2'
    labelsx.append(label)
```

```python
plt.figure(figsize=(12,8))
plt.rcParams.update({'font.size': 22})

plt.plot(x,yhidden,'-',label = 'Hidden variables')
plt.plot(x,ystat, '-' ,label = 'Real values')

plt.grid(which='major',axis='both')
plt.rcParams.update({'font.size': 16})
plt.legend(loc='upper left')

plt.xticks(stepsx, labelsx)
plt.xlabel(r'$\theta$ (rad)')
plt.ylabel('Correlation')
plt.plot()
```
